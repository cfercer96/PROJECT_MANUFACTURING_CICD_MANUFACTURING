Permisos a los schemas y catalogs

```sql
GRANT ALL PRIVILEGES ON CATALOG catalog_project TO `data_engineers`;
GRANT USAGE ON CATALOG catalog_project TO `bi_readers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  catalog_project.bronze to `data_engineers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  catalog_project.silver to `data_engineers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  catalog_project.gold to `data_engineers`;

GRANT USAGE ON CATALOG catalog_project TO `bi_readers`;
GRANT USAGE ON SCHEMA catalog_project.gold TO `bi_readers`;
GRANT SELECT ON SCHEMA catalog_project.gold TO `bi_readers`;


```
